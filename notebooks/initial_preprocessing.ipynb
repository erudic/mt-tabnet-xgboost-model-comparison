### Weather_Condition

```python
df['Weather_Condition'].value_counts()
```

```python
weathers = df['Weather_Condition'].value_counts().index
[w for w in weathers if 'Windy' in w]
```

Trying to reduce number of classes. Checking that windy present means that Wind_Direction is also present ie. it is not calm.

```python
contain_windy_and_calm = df['Weather_Condition'].str.contains('Windy',na=False) & df['Wind_Direction'].str.contains('Calm', na=False)
df[contain_windy_and_calm][['Weather_Condition','Wind_Direction']]
```

We can see that wind direction is present everytime the weather is windy. So the **Windy** weather condition will be removed. 

Also checking how many weather types contain multiple descriptions.

```python
[w for w in weathers if '/' in w]
```

Splitting weathers into an array and removing windy weather condition.

```python
def map_weather_to_arr(weather_str):
    return [weather.strip() for weather in weather_str.split(' / ')]

def remove_windy(weather_arr):
    try:
        weather_arr.remove('Windy')
    except ValueError as e:
        pass
    return weather_arr

df['Weather_Condition_Arr']=df['Weather_Condition'].map(map_weather_to_arr,na_action='ignore').map(remove_windy,na_action='ignore')
preprocessed_columns = ['Weather_Condition']
```

Mostly additional weather descriptions are windy. Exceptions are: **Dust Whirlwinds**, **Dust Whirls Nearby** and **Wintry Mix**. We'll check actually how many of those conditions are labeled in *preprocessing* notebook.

## Encoding wind direction

```python
def map_wind(wind_str, neg_direction,pos_direction):
    if wind_str=='Calm':
        return 0
    
    n = len(wind_str)
    s = 0
    for char in wind_str:
        char_val = 0
        if char==pos_direction:
            char_val = 1
        if char==neg_direction:
            char_val = -1
        s += char_val
    return s/n

map_sn = lambda wind_str: map_wind(wind_str,'S','N')
map_ew = lambda wind_str: map_wind(wind_str,'E','W')

df['Wind_SN'] = df['Wind_Direction'].map(map_sn, na_action='ignore')
df['Wind_EW'] = df['Wind_Direction'].map(map_ew, na_action='ignore')
df[['Wind_SN','Wind_EW']]
winds = df[['Wind_SN','Wind_EW']]
encoded_cols = ['Wind_Direction']
winds
```

## Feature selection

```python
drop_cols = useless_columns + high_card_cols + out_of_scope_cols + encoded_cols + preprocessed_columns
df = df.drop(drop_cols, axis='columns')
```
