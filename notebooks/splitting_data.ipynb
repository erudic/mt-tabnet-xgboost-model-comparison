```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('../data/dataset_initial.csv')

df_train_val,df_test = train_test_split(df, train_size=0.8, stratify=df['Severity'])
```

```python
df_train_val.to_csv('../data/large/train_val.csv',index=False)
df_test.to_csv('../data/large/test.csv', index=False)
```
